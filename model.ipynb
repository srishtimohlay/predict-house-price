```python
house_processed=housing
cat_cols_fill_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                     'GarageCond', 'GarageQual', 'GarageFinish', 'GarageType',
                     'BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtQual', 'BsmtCond',
                     'MasVnrType']
for col in cat_cols_fill_none:
    house_processed[col]= house_processed[col].fillna("None")

    
house_processed['LotFrontage']= house_processed.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))    

for col in ['GarageYrBlt', 'GarageArea', 'GarageCars', 'MasVnrArea']:
    house_processed[col] = house_processed[col].fillna(int(0))

house_processed['Electrical'] = house_processed['Electrical'].fillna(house_processed['Electrical']).mode()[0]

house_processed = house_processed.drop(['Utilities'], axis=1)

house_processed.isna().sum()
```

```python
num_attributes = house_processed.select_dtypes(exclude='object')

high_quant = house_processed.quantile(.999)
for i in num_attributes.columns:
    house_processed = house_processed.drop(house_processed[i][house_processed[i]>high_quant[i]].index)
```

```python
attributes_drop = ['MiscVal', 'MoSold', 'YrSold', 'BsmtFinSF2','BsmtHalfBath','MSSubClass',
                   'GarageArea', 'GarageYrBlt', '3SsnPorch']

house_processed = house_processed.drop(attributes_drop, axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_processed_1hot = cat_encoder.fit_transform(house_processed)
housing_processed_1hot
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

X= house_processed.drop("SalePrice", axis=1)
y= house_processed["SalePrice"].copy()
get_numeric_data= X.select_dtypes(exclude='object')
get_cat_data= X.select_dtypes(include='object')

num_attribs = list(get_numeric_data)
cat_attribs = list(get_cat_data)

num_pipeline= Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('scaler', StandardScaler())
])

pipe= ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(),cat_attribs)
])
    

print(y.describe())

print(X)

y  = np.log(y)

X = pipe.fit_transform(X)

print(y.describe())

print(X)


```

```python
#spiliting data to train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=7)

from sklearn.metrics import mean_squared_error as MSE

def toexp(y):
    return np.exp(y)

rmse = pd.Series()
rmse.index.name = 'Model'
scores = pd.Series()
scores.index.name = 'Model'

def rmse1(name, y_pred):
    rmse[name]= MSE(toexp(y_pred), toexp(y_test))**(1/2)
    
def score1(name, model):
    score[name]= model.score(X_test, y_test)
      
```

```python
#Regression models

#Model 1: Linear Regression.
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr= lr.predict(X_test)
rmse1('Linear Regression', y_pred_lr)
score1('Linear Regression', lr)


#Model 3: Ridge Regression.
from sklearn.linear_model import Ridge
ri=Ridge()
ri.fit(X_train, y_train)
y_pred_ri=ri.predict(X_test)
rmse1('Ridge Regression', y_pred_ri)
score1('Ridge Regression', ri)


#Model 4: Lasso Regression.
from sklearn.linear_model import Lasso
la= Lasso()
la.fit(X_train, y_train)
y_pred_la= la.predict(X_test)
rmse1('Lasso Regression', y_pred_la)
score1('Lasso Regression', la)


#Model 5: Elastic Net.
from sklearn.linear_model import ElasticNet
el = ElasticNet()
el.fit(X_train, y_train)
y_pred_el= el.predict(X_test)

rmse1('Elastic Net Regression', y_pred_el)
score1('Elastic Net Regression', el)



#Model 6: Random forest Regression.
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf= rf.predict(X_test)

rmse1('Random forest', y_pred_rf)
score1('Random forest', rf)


#Model 6: Decision Tree Regression.
from sklearn.tree import DecisionTreeRegressor
dt= DecisionTreeRegressor()
dt.fit(X_train, y_train)
y_pred_dt= dt.predict(X_test)

rmse1('Decision Tree', y_pred_dt)
score1('Decision Tree', dt)
```

```python
from sklearn.model_selection import cross_val_score

# Perform K fold cross-validation, where K=10
scores = cross_val_score(lr, X_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
linear_rmse_scores = np.sqrt(-scores)

# Display results
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(linear_rmse_scores)
```
